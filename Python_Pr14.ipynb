````python
from statistics import *
import csv
import json
from collections import defaultdict
from datetime import datetime

def get_quart(date):
    month = datetime.strptime(date,"%d-%m-%Y").month
    if 1 <= month <= 3:
        return 'Q1'
    if 4 <= month <= 6:
        return 'Q2'
    if 7 <= month <= 9:
        return 'Q3'
    if 10 <= month <= 12:
        return 'Q4'
        
with open('grades.json') as f:
    data = json.load(f)

dict_by_group = defaultdict(list)

for entry in data: #{'name': 'Bob', 'subject': 'Science', 'grade': 86, 'date': '06-09-2025'}
    quart = get_quart(entry['date'])
    if quart:
        key = (entry['subject'], entry['name'], quart)
        dict_by_group[key].append(entry['grade'])

grades_by_group = [dict([items]) for items in dict_by_group.items()]
sorted_by_subject = sorted(grades_by_group, key=lambda x: list(x.keys())[0][0])
results = []
for (subject, name, q), grades in sorted_by_subject[0].items(): #[{('History', 'Ethan', 'Q2'): [68]},...
    avg_grades = mean(grades)
    med_grades = median(grades)
    results.append([subject, name, q, avg_grades, med_grades])

with open('quarterly_report.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(['Subject','Name','Quarter','Average grade','Median grade'])
    writer.writerows(results)
````
