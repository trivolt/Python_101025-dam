4. Напишите программу, которая получит от пользователя возраст и выведет соответствующую категорию: ребенок (меньше 12 лет), подросток (от 12 до 18), взрослый (от 18 до 60), пожилой (больше 60).  
Пример вывода:  
Введите возраст: 25  
Вы взрослый.  


````python
age = int(input('Введите возраст: '))

if age < 0:
    print('Введено не корректное значение возраста')
else:
    if age < 12:
        print('ребенок')
    elif age < 18:
        print('подросток')
    elif age < 60:
        print('взрослый')
    else:
        print('пожилой')
        
````

````python
v = int(input('Введите возраст: '))

match v:
    case _ if v < 12:
        print("Ребенок")
    case _ if 12 <= v < 18:
        print("Подросток")
    case _ if 18 <= v < 60:
        print("Взрослый")
    case _ if v >= 60:
        print("Пожилой")
````

9.* Напишите программу, которая принимает три числа и определяет знак их произведения без явного умножения.  
Пример вывода:  
Введите три числа:  
-2  
5  
-3  
Произведение положительное.  


````python
a = int(input('Введите первое число:'))
b = int(input('Введите второе число:'))
c = int(input('Введите третье число:'))

negative_count = (a<0) + (b<0) + (c<0)

if a == 0 or b == 0 or c == 0:
    print('Произведение нулевое')
else:    
    if negative_count % 2 == 0:
        print('Произведение положительное.')
    else:
        print('Произведение отрицательное.')
````
