````python
import csv
import json
from statistics import *
from collections import defaultdict
from datetime import datetime

def get_stats(data: dict) -> dict:
    result = {}

    for key, items in data.items():
        revenues = [item['revenue'] for item in items]
        dates = [item['date'] for item in items]

        if not revenues or not dates:
            result[key] = {
                "total_revenue": sum(revenues),
                "average_revenue": mean(revenues),
                "median_revenue": median(revenues),
                "stddev_revenue": round(stdev(revenues), 2) if len(revenues) > 1 else 0,
                "first_sale_date": min(dates).strftime("%Y-%m-%d"),
                "last_sale_date": max(dates).strftime("%Y-%m-%d")
            }

    return result

records = []
fruit_data = defaultdict(list)
store_data = defaultdict(list)

with open('sales.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        fruit = row['Fruit']
        store = row['Store']
        date = datetime.strptime(row['Date'], "%Y-%d-%m")
        revenue = int(row['Quantity'])*float(row['Price'])

        fruit_data[fruit].append({
            'date': date,
            'revenue' : revenue
        }) 
        store_data[store].append({
            'date': date,
            'revenue' : revenue
        })

fruit_stats = get_stats(fruit_data)
store_stats = get_stats(store_data)

with open('fruit_summary.json', 'w') as f:
    json.dump(fruit_stats, f, indent=3)
    
with open('store_summary.json', 'w') as f:
    json.dump(store_stats, f, indent=3)

print('Смотри результат в папке CSV')
````
