Напишите рекурсивную функцию fib(n), которая возвращает n-е число Фибоначчи.

````python
# F(n) = F(n-1)+F(n-2)

def f(n):
    if n==1:
        return 0
    if n==2:
        return 1
    return f(n-1) + f(n-2)

f(5)
````

Напишите сумму чисел от 1 до n рекурсивно

````python
def s(n):
    if n==1:
        return 1
    return n + s(n-1)

print(s(5))
````

Напишите рекурсивную функцию is_palindrome(s), которая проверяет, является ли строка s палиндромом (читается одинаково слева направо и справа налево).

````python
def is_palindrome(s):
    if len(s)<=1:
        return True
    return s[0]==s[-1] and is_palindrome(s[1:-1])

print(is_palindrome('abba'))
````

Генерация всех подмножеств print(subsets([1, 2]))  # [[], [2], [1], [1, 2]]

````python
def subsets(nums):
    if not nums:
        return [[]]
    rest = subsets(nums[1:])
    return rest + [[nums[0]] + subset  for subset in rest]
    

nums = [1,2]
subsets(nums)
````

````python
def count_dig(n):
    n=abs(n)
    if n<10:
        return 1 if n%2==0 else 0
    return (1 if (n%10)%2==0 else 0) + count_dig(n//10)

print(count_dig(-1234))
````

2. ***Контроль ввода***  
Пользователь вводит строку с информацией о студенте: фамилия, имя, год рождения, курс – через пробел. Напишите функцию, которая проверяет корректность ввода: первые два элемента должны быть строками (из букв, без лишних символов), вторые два элемента – целые числа в подходящем диапазоне. В случае нарушения формата функция должна выбрасывать ошибку. В случае корректного ввода функция возвращает словарь с информацией о студенте. При вызове функции обработайте исключение.

*Пример ввода 1:*   
Введите информацию о студенте (фамилия имя год_рождения курс):  Smith John 2000 3  

*Пример вывода:*   
Введенная информация корректна: `{'last_name': 'Smith', 'first_name': 'John', 'birth_year': 2000, 'course': 3}`  

*Пример ввода 2:*   
Введите информацию о студенте (фамилия имя год_рождения курс):  Smith Jo.hn 2000 3  

*Пример вывода:*   
Ошибка ввода: Имя должно содержать только буквы. 


````python
from datetime import datetime

datetime.now().year

````

````python
from datetime import datetime

def transform_to_dict(s:str):
    split_info = s.split()
    if len(split_info)!=4:
        raise ValueError('Неверное количество аргументов')
    l_name, f_name, birth, course = split_info
    if not l_name.isalpha():
        raise  ValueError('Фамилия должна состоять из букв')
    if not f_name.isalpha():
        raise  ValueError('Имя должно состоять из букв')
    if not birth.isdigit():
        raise  ValueError('Год рождения должен быть числом')
    if not 1900<=int(birth)<=datetime.now().year - 10:
        raise  ValueError('Неверный интервал года рождения')
    if not course.isdigit():
        raise  ValueError('Курс должен быть числом')
    if not 1<=int(course)<=6:
        raise  ValueError('Курс имеет неверный диапазон')
    d = {}
    d['last_name'] = l_name
    d['first_name'] = f_name
    d['birth'] = birth
    d['course'] = course
    return d


transform_to_dict('Smith John 2000 3')
````

````python
try:
    print(transform_to_dict('Smith John 2000 13'))
except ValueError as e:
    print(f'Ошибка ввода: {e}')
    

````

3. ***Сумма чисел списка***  
Напишите рекурсивную функцию, которая вычисляет сумму всех чисел в списке.  
Функция должна проверять:  
Аргумент должен быть списком.  
Все элементы списка должны быть числами.  
Если данные не валидны необходимо выбрасывать исключение. При вызове функции обработайте возможное исключение.

*Данные:*   
`numbers = [1, 2, 3, 4, 5]`  
Пример вывода:   
15


````python
def rec_sum(lst:list):
    if not isinstance(lst, list):
        raise TypeError('Ожидается список чисел')
    if not isinstance(lst[0],(int,float)):
        raise ValueError('Все элементы списка должны быть числом')
    if not lst:
        return 0
    return lst[0] + rec_sum(lst[1:])


numbers = [1, 2, 'dgh', 4, 5]
try:
    print(rec_sum(numbers))
except (TypeError, ValueError) as e:
    print('Ошибка:',e)
    
````

4. ***Реверс строки***  
Напишите рекурсивную функцию, которая переворачивает строку. Если передан не строковый тип, выбросить исключение. При вызове функции обработайте возможное исключение.

*Данные:*   
`text = "recursion"`
*Пример вывода:*   
noisrucer


````python
def reverse_text(text):
    if not isinstance(text, str):
        raise TypeError('Ожидается строка')
    if len(text) <= 1:
        return text
    return text[-1] + reverse_text(text[:-1])

try:
    text = "recursion"
    print(reverse_text(text))
except TypeError as e:
    print('Ошибка:', e)
    
````

5. ***Глубина вложенности списка***  
Напишите рекурсивную функцию, которая определяет максимальную глубину вложенности списка. Функция должна проверять:  
Аргумент должен быть списком.  
Вложенные структуры у которых проверяется глубина, – тоже списки. Глубина остальных вложенных структур и объектов считается 1.  


*Данные:*   
`nested_list = [1, [2, [3, [4, [5]]]], 6, [[7, 8], 9]]`  

*Пример вывода:*   
Максимальная глубина: 5


````python
def max_depth(lst)->int:
    if not isinstance(lst,list):
        raise TypeError('Ожидается вложенный список')
        
    # Базовый случай: если в списке нет других списков
    if not any(isinstance(item, list) for item in lst):
        return 1
    #Рекурсивный случай
    depth = [max_depth(item)  for item in lst if isinstance(item, list)]
    return 1 + max(depth)

nested_list = [1, [2, [3, [4, [5]]]], 6, [[7, 8], 9]]
try:
    print(f'Максимальная глубина: {max_depth(nested_list)}')
except TypeError as e:
    print('Ошибка:',e)
    
````

````python
def sum_sales(department):
    if not isinstance(department,dict):
        raise TypeError('Ожидается словарь с информацией о департаменте')
    total = department['sales'] #можно проверить на наличие ключа и его типа, но мы не будем)))
    for sub_dept in department.get('sub_departments',[]):
        total += sum_sales(sub_dept)
    return total
````
