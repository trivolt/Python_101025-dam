2. ***Частотный анализ букв***  
Напишите программу, которая подсчитывает количество вхождений каждой буквы в тексте.  
Программа должна игнорировать регистр слов, пробелы и символы "." и ",".  
* Можно решить более сложную задачу и игнорировать все небуквенные символы.  
*Данные:*  
`text = "This is a test. This test is only a test."`  
*Пример вывода:*  
`{'t': 8, 's': 7, 'i': 4, 'e': 3, 'h': 2, 'a': 2, 'o': 1, 'n': 1, 'l': 1, 'y': 1}`


````python
from collections import Counter

text = "This is a test. This test is only a test."

def counter_alp(text):
    cnt_alp = Counter(text.lower().replace('.','').replace(',','').replace(' ',''))
    return cnt_alp.most_common()
    
print(dict(counter_alp(text)))
````

````python
from collections import Counter

text = "This is a test. This test is only a test."

def counter_alp(text):
    return Counter([ch for ch in text.lower() if ch.isalpha()]).most_common()
    
print(dict(counter_alp(text)))
````

3. ***Популярные слова***  
Реализуйте функцию, которая принимает любое количество строк с текстом.  
Функция должна возвращать подсчет самых популярных слов в количестве, переданном в функцию. Программа должна игнорировать регистр слов. Выведите 5 самых популярных слов и их количество.  
*Данные:*  
`text1 = "This is a sample text with some repeated words."`  
`text2 = "Another sample text with different words."`  
`text3 = "Text processing is fun when words repeat."`  
*Пример вывода:*  
5 самых популярных слов:  
```
	words: 3
	text: 3
	with: 2
	sample: 2
	is: 2
```

````python
text1 = "This is a sample text with some repeated words."
text2 = "Another sample text with different words."
text3 = "Text processing is fun when words repeat."

def clean_text(text):
    return text.lower().replace('.','').replace(',','')

def print_top_words(dct):
    print('5 самых популярных слов:')
    for word, count in dct.items():
        print(f'{word}: {count}')
    

def count_words(*args):
    t1 = Counter('')
    for text in args:
        t1 += Counter(clean_text(text).split())
    return dict(t1.most_common(5))

print_top_words(count_words(text1, text2, text3))
````

4. ***Группировка задач по категории***  
Реализуйте функцию, которая принимает словарь задач с категориями и группирует задачи по их категориям.  
*Данные:*  
```
tasks = {
    "task1": "работа",
    "task2": "учёба",
    "task3": "развлечения",
    "task4": "работа",
    "task5": "учёба"
}
```
*Пример вывода:*   
Группировка по категориям:  
```
{
    'работа': ['task1', 'task4'],
    'учёба': ['task2', 'task5'],
    'развлечения': ['task3']
}
```

````python
from collections import defaultdict

tasks = {
    "task1": "работа",
    "task2": "учёба",
    "task3": "развлечения",
    "task4": "работа",
    "task5": "учёба"
}

def grouped_work(tasks):
    dd = defaultdict(list)
    for task, work in tasks.items():
        dd[work].append(task)
    return dict(dd)

def print_tasks(dct):
    print('Группировка по категориям:')
    print(dct)

print_tasks(grouped_work(tasks))
````

6. ***Группировка слов по длине***  
Реализуйте функцию, которая группирует слова по их длине и возвращает словарь с ними.   
*Данные:*  
`words = ["apple", "banana", "kiwi", "grape", "orange", "peach"]`  

*Пример вывода:*
Слова по длине:
```
5: ['apple', 'grape', 'peach']
6: ['banana', 'orange']
4: ['kiwi']
```

````python
from collections import defaultdict

words = ["apple", "banana", "kiwi", "grape", "orange", "peach"]

def group_by_len_word(words):
    dict_len_words = defaultdict(list)
    for word in words:
       dict_len_words[len(word)].append(word)
    return dict(dict_len_words)

print('Слова по длине:')
for l, fruits in group_by_len_word(words).items():
    print(l, fruits)

````

7. ***Создание глобального счётчика***  
Создайте функцию increment_counter, которая сообщает, что была вызвана, и использует глобальную переменную counter для подсчёта числа своих вызовов.
`Пример вызова:`
```
increment_counter()
increment_counter()
print(counter)
```
*Пример вывода:*  
Вызовов функции: 2


````python
cnt = 0

def increment_counter():
    global cnt
    cnt += 1
    return cnt

for _ in range(10000):
    increment_counter()

print(cnt)
````

8. ***Замер времени работы программы.***  
Напишите программу, которая выводит все делители заданного пользователем числа и сообщает, за какое время была решена эта задача. Используйте модуль time.
Пример ввода и вывода:  
Введите число: 54832  
Делители числа: `[1, 2, 4, 8, 16, 23, 46, 92, 149, 184, 298, 368, 596, 1192, 2384, 3427, 6854, 13708, 27416, 54832]`  
Время выполнения: `0.005539` секунд


````python
import time

def counter_div(n):
    return [i  for i in range(1,n+1) if n % i == 0]


n = 54832
start = time.time()

print(f'Делители числа: {counter_div(n)}')

end = time.time()

print(f'Время выполнения: {end - start:.6f} секунд')

````
